# Distribution of island sizes

Does the area of world islands follow a power law?  What if you include the continents?

```r
library(ggplot2)
library(dplyr)
library(reshape2)

# Run from a clone of the git repo https://github.com/rgiordan/IslandDistribution.

# Get the base directory:
git_repo_dir <- system("git rev-parse --show-toplevel", inter=TRUE)
data_dir <- file.path(git_repo_dir, "data")
```

```r
ReadTable <- function(table_name) {
    d <- read.csv(file.path(data_dir, table_name), stringsAsFactors=FALSE)

    # Convert the area column to a numeric value.
    # The area column has slightly different names in the differnet tables
    area_df <- select(d, starts_with("Area.km2"))
    stopifnot(ncol(area_df) == 1)
    area_text <- as.character(area_df[, 1])

    # Strip out commas and footnotes and change to a number
    area_numeric <- as.numeric(gsub("\\[.*\\]", "", gsub(",", "", area_text)))
    stopifnot(all(is.finite(area_numeric)))
    
    d[["area_km2"]] <- area_numeric
    d[["table"]] <- sub("\\.csv$", "", table_name)
    return(d)
}
```

Load the data into a single dataframe.

```r
data_files <- c("continents.csv", sprintf("islands_%d.csv", 2:7))
df_list <- lapply(data_files, ReadTable)

print(df_list[[1]])

areas_df <-
   do.call(rbind, lapply(df_list, function(x) { select(x, area_km2, table) })) %>%
   arrange(area_km2) %>%
   mutate(rank=1:n() / n(), continent=grepl("continents", table))
```

Obviously most islands are small.

```r
ggplot(areas_df) +
    geom_point(aes(x=rank, y=area_km2, color=continent), size=2) +
    ggtitle("All islands")
```

If a density follows a power law, so does its cumulative distribution function.

$$
f(x) \propto x^{-(1+\alpha)} \Rightarrow\\
P(x < x_0) = \int_{x_{min}}^{x_0} f(x) dx \propto \frac{x_0^{-\alpha}}{-\alpha}
$$

So if the island sizes follow a power law, then the log-log plot of the ranks and island sizes should be a straight line.

## Results

Evidently all the islands taken together do not follow a power law.

```r
ggplot(areas_df) +
    geom_point(aes(x=rank, y=area_km2, color=continent), size=2) +
    scale_y_log10() + scale_x_log10() +
    ggtitle("All islands")
```

Perhaps the areas of smaller islands are misreported -- only look at bigger islands.

Even the largest islands are still bigger than would be predicted by a pure power law.  The continents seem to continue a trend that is consistent with the largest islands, though.

```r
ggplot(filter(areas_df, rank > 0.8)) +
    geom_point(aes(x=rank, y=area_km2, color=continent), size=2) +
    scale_x_log10() + scale_y_log10()
```
